### Intro to Matplotlib
Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. It is used along with
NumPy to provide an environment that is an effective open source alternative to MatLab. Thus, this may feel very familiar to you if
you've ever used MatLab. Please note that Matplotlib is not the only tool for creating data visualization in Python. Other common
libraries include seaborn, plotly, and even pandas has some built in functions for quick plotting. This lab will mainly focus on
Matplotlib, but I highly recommend checking out seaborn as well, since it can make some type of plots much easier to create. The
good news is that seaborn is built on top of Matplotlib, so it will be very easy to pick up after going through this lab. However,
you're free to use any of the plotting tools mentioned above throughout the course. Just as there were many ways to solve a
problem in pandas, there are many ways to visualize our data.
### Let's begin by importing matplotlib and learning how to create some beautiful graphs. If for some reason you don't have matplotlib installed, you will first have to go to your terminal (or Anaconda Prompt if on Windows) and enter the following:

```python
""" Import all the necessary libraries and then set grid displays """

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams

rcParams['figure.figsize'] = 15 , 5
sns.set_style('darkgrid')
```

#### Plotting Numpy Arrays

```python
""" Plotting Numpy arrays """ 
x = np.linspace(0,20)
y = np.sin(x)
plt.plot(x,y)
plt.show()

plt.plot(x,y)
plt.title('My Graph Title')
plt.xlabel('My x-axis label')
plt.ylabel('My y-axis label')
plt.show()

plt.plot(x, np.sin(x), label = 'y=sin(x)')
plt.plot(x, np.cos(x), label = 'y=cos(x)')
plt.title('Sine and Cosine Functions')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.show() 
```

### Different examples

```python
""" Different examples """
x = np.linspace(0,40)
y = np.sin(x)
plt.plot(x,y)
plt.show()

plt.plot(x,y)
plt.title('My Graph Title example')
plt.xlabel('My x-axis label')
plt.ylabel('My y-axis label')
plt.show()

plt.plot(x, np.sin(x), label = 'y=sin(x)')
plt.plot(x, np.tan(x), label = 'y=tan(x)')
plt.title('Sine and Tan Functions')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.show() 
```

#### Subplot

```python
#one figure object with 1 x 2  = 2 axes
x = np.linspace(0,20)
fig, axes = plt.subplots(nrows = 1, ncols= 2)
#plot sin(x) in axes[0] left plot 
axes[0].plot(x, np.sin(x))
axes[0].set_title('y = sin(x)')

#plot for cos(x) in axes[1] right plot
axes[1].plot(x, np.cos(x))
axes[1].set_title('y = cos(x)')
```

### Custom Axes range

```python
x=np.linspace(0,5,11)
 
fig,axes=plt.subplots(nrows=1,ncols=3)
 
#wecanalsoplotmultiplegraphsusingplt.plotx1,y1,x2,y2.
axes[0].plot(x,x**2,x,x**4)
axes[0].set_title('defaultaxisrange')
 
axes[1].plot(x,x**2,x,x**4)
#tighthappenstobethesamehere,butthiswillcomeinhandywhenyouwanttoautomaticallyalignaxesi
#fittedmanner.
axes[1].axis('tight')
axes[1].set_title('tightaxisrange')
 
#noticethezoominthathappened.
axes[2].plot(x,x**2,x,x**4)
axes[2].set_ylim([0,40])
axes[2].set_xlim([2,3])
axes[2].set_title("customaxesrange")
 
plt.show()
```

### Stackplot

```python
x = pokemon_df['Name'][:4]
y1 = pokemon_df['HP'][:4]
y2 = pokemon_df['Attack'][:4]
y3 = pokemon_df['Defense'][:4]

labels = ["HP", "Attack", "Defense"]
plt.stackplot(x,y1,y2,y3)
plt.legend(loc='upper left', labels=labels)
plt.xticks(rotation=90)
plt.show()

fig, axes = plt.subplots(nrows=3, ncols=1, sharey=True)
axes[0].plot(x ,y1, 'ko-')
axes[0].set(title='3 subplots', ylabel='HP')
axes[1].plot(x ,y2, 'r.-')
axes[1].set(xlabel='Pokemon', ylabel='Attack')
axes[2].plot(x ,y3, ':')
axes[2].set(xlabel='Pokemon', ylabel='Defense')
plt.show()

plt.plot(pokemon_df['Attack'], label='Attack')
plt.plot(pokemon_df['HP'], label='HP')
plt.legend()
plt.show()
```
